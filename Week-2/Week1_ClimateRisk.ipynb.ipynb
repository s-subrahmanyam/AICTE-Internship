```python
# -----------------------------------------------
# Week 1 Project - Climate Risk and Disaster Management
# Dataset: World Disaster Risk Index Time Series
# -----------------------------------------------

# 1. Import necessary libraries
# pandas: for data manipulation
# numpy: for numerical operations
# matplotlib.pyplot: for creating visualizations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 2. Load the dataset
# Make sure the CSV file is in the same folder as this notebook/script
data_file = "global_disaster_risk_index_time_series.csv"
data = pd.read_csv(data_file)

# Remove any leading/trailing spaces in column names
# This avoids errors when accessing columns like ' Lack of Adaptive Capacities'
data.columns = data.columns.str.strip()

# 3. Explore the dataset

# Display first 5 rows to see sample data
print("----- First 5 Rows -----")
print(data.head())

# Display dataset information: number of entries, column names, data types
print("\n----- Dataset Info -----")
print(data.info())

# Display statistical summary for numeric columns (mean, min, max, quartiles)
print("\n----- Dataset Description -----")
print(data.describe())

# Check for missing values in each column
print("\n----- Missing Values -----")
print(data.isnull().sum())

# 4. Visualizations

# Histogram of World Risk Index (WRI)
# Shows the distribution of risk levels across countries
if 'WRI' in data.columns:
    plt.figure(figsize=(8,5))
    plt.hist(data['WRI'], bins=30, color='skyblue', edgecolor='black')
    plt.title('Distribution of World Risk Index (WRI)')
    plt.xlabel('WRI')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
else:
    print("Column 'WRI' not found in dataset. Columns:", data.columns)

# Time Series Plot: Average WRI per Year
# Shows how global disaster risk changes over time
if 'Year' in data.columns and 'WRI' in data.columns:
    yearly_wri = data.groupby('Year')['WRI'].mean()  # Calculate average WRI per year
    plt.figure(figsize=(10,6))
    plt.plot(yearly_wri.index, yearly_wri.values, marker='o', linestyle='-')
    plt.title('Average World Risk Index Over Years')
    plt.xlabel('Year')
    plt.ylabel('Average WRI')
    plt.grid(True)
    plt.show()
else:
    print("Columns 'Year' or 'WRI' not found. Columns:", data.columns)

# 5. Improvements / Highlights implemented:
# - Fixed column name issues (stripped spaces)
# - Replaced 'RiskIndex' with 'WRI' to match dataset
# - Added missing value checks
# - Added histogram and time series plots for visualization
# - Grouped WRI by year to show trend over time
# -----------------------------------------------
```
